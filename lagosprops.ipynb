```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

<h3><h2>Code to Scrape Pages online<h1>

```python

proceed = True
current_page = 1
data = []
while (proceed):

    print(f"Currently scrapping page {current_page}")
    url = f'https://nigeriapropertycentre.com/for-rent/lagos?page={current_page}'
    response = requests.get(url).text
    soup = BeautifulSoup(response, 'html.parser')
    #print(soup.prettify())

    if current_page >= 1081:
        proceed = False
    else:
        listings = soup.find_all('div', class_='wp-block property list')
        for listing in listings:
            item = {}
            item['desc'] = listing.find('h4', class_= 'content-title').text.strip()
            item['address'] = listing.find('address', class_= 'voffset-bottom-10').text.strip()
            item['link'] = 'https://nigeriapropertycentre.com/' + listing.find('a').attrs['href']
            item['price'] = listing.find('span', class_= 'pull-sm-left').text.strip()
            attributes = [i.text.strip() for i in listing.find('ul', class_='aux-info')]
            
            #assign scraped features to dictionary item
            if len(attributes) == 1:
                item['bedrooms'] = attributes[0]
                item['bathrooms'] = 'Not Stated'
                item['toilets'] = 'Not Stated'
                item['parking space'] = 'Not Stated'
            elif  len(attributes)==2:
                item['bedrooms'] = attributes[0]
                item['bathrooms'] = attributes[1]
                item['toilets'] = 'Not Stated'
                item['parking space'] = 'Not Stated'
            elif  len(attributes)==3:
                item['bedrooms'] = attributes[0]
                item['bathrooms'] = attributes[1]
                item['toilets'] = attributes[2]
            elif  len(attributes)>=4:
                item['bedrooms'] = attributes[0]
                item['bathrooms'] = attributes[1]
                item['toilets'] = attributes[2]
                if attributes[3] == 'save':
                    item['parking space'] = 'Not Stated'
                else:
                    item['parking space'] = attributes[3]
                
            data.append(item) 
    current_page += 1

    #a dataframe df is created to hold all scraped data
df = pd.DataFrame(data) 
df.head(20)
   

```

```python
#It is imperative to add a location column that picks it value(town in Lagos) from the address column
df['location'] = df['address'].apply(lambda x:  x.split()[-2])

df['location']
```

```python
df['location'] = df['location'].str.replace(',', '')
df['location']
```
