```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://nigeriapropertycentre.com/for-sale')
html_page = response.text
```

```python
soup = BeautifulSoup(html_page, 'html.parser')
soup
```

```python
properties = soup.find_all('div', class_ = 'wp-block-body')

link = []
Address = []
Label = []
Price = []
for property in properties:
    link.append(property.a['href'])
    Label.append(property.find('h4', class_= 'content-title').text)
    Address.append(property.find('address', class_= 'voffset-bottom-10').text)
    Price.append(property.find('span', class_= 'pull-sm-left').text.strip())
```

```python
len(Address)
```

```python

attributes = soup.find_all('ul', class_ = 'aux-info') 

Rooms = []
Bathrooms = []
Toilets = []
Parking = []

for attribute in attributes:
   
    span_tags = attribute.find_all('span')
    Rooms.append(span_tags[0].text)
    Bathrooms.append(span_tags[2].text)
    Toilets.append(span_tags[4].text)       

    
```

```python
property_details = pd.DataFrame()
property_details['Label'] = Label
property_details['Address'] = Address
property_details['link'] = link
property_details['Rooms'] = Rooms
property_details['Toilets'] = Toilets
property_details['Bathrooms'] = Bathrooms
property_details['Price'] = Price
```
